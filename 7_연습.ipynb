### 강의자료 1 문제

```python
import keyboard, time
import datetime as dt
import os

while True:
    k=input('게임을 시작하려면 s키를 누르세요')
    if k=='s':
        pass_t=dt.datetime.now()
        break
    else:
        print('press s')

while True:
    print('10초가 지났다고 생각되면 press q')
    os.system('cls')
    if keyboard.is_pressed('q'):
        break
    
cur_t=dt.datetime.now()

time_gap=(cur_t.minute*60+cur_t.second)-(pass_t.minute*60+pass_t.second)

if int(time_gap)==10:
    print('시간 감각이 뛰어나네요')
else:
    print(time_gap)
    print('시간 감각이 꽝이에요')
```

```python
from tkinter import *
```

```python
def change_label():
    if label.cget('text')=='헬로 파이썬':
        label.config(text='Hi Python')
        label.config(bg='cyan')
    else:
        label.config(text='헬로 파이썬')
        label.config(bg='yellow')
        
window=Tk()
label=Label(window,text='헬로 파이썬', bg='yellow')
label.pack()
btn=Button(window, text='클릭하면 문자가 변경됨',fg='blue',command=change_label)
btn.pack()
window.mainloop()
```

```python
def entry_to_label():
    str=input_entry.get()
    label.config(text=str)
```

```python
window=Tk()

input_entry=Entry(window,width=50)
input_entry.pack()

button=Button(window, text='처리',command=entry_to_label)
button.pack()

label=Label(window,text=' ')
label.pack()

window.mainloop()
```

```python
from tkinter import *
def bmi_compute():
    w=float(weight.get())
    h=float(height.get())/100.0
    bmi=w/(h*h)
    answer='당신의 체질량 지수(BMI)는: {:4.2f}'.format(bmi)
    result.config(text=answer)
    
window=Tk()
label=Label(window, text='체중(kg)과 키(cm)를 차례로 입력하세요.')
label.pack()
weight=Entry(window,width=50); weight.pack()
height=Entry(window,width=50); height.pack()
button=Button(window, text='BMI 계산', command=bmi_compute); button.pack()
result=Label(window, text='당신의 체질량 지수(BMI)는:'); result.pack()
window.mainloop()
```

```python
from tkinter import *
window=Tk()
window.title('사칙연산 계산기')
window.geometry('350x200')
Label(window, text='숫자 1',font=('20')).grid(row=0, column=0)
Label(window, text='숫자 2').grid(row=1, column=0)

res_label=Label(window, text='결과 :')
res_label.grid(row=2, column=0)
result=Label(window, text='')
result.grid(row=2, column=1)

n1=Entry(window, width=10)
n2=Entry(window, width=10)
n1.grid(row=0, column=1)
n2.grid(row=1, column=1)

def add():
    res_text=str(float(n1.get())+float(n2.get()))
    result.config(text=res_text)
    
def subtract():
    res_text=str(float(n1.get())-float(n2.get()))
    result.config(text=res_text)

def multiplication():
    res_text=str(float(n1.get())*float(n2.get()))
    result.config(text=res_text)
    
def division():
    res_text=str(float(n1.get())/float(n2.get()))
    result.config(text=res_text)
    
def era():
    n1.delete(0,END); n2.delete(0,END); result.configure(text='')

btn_clear=Button(window, text='지우기', width=3, command=era)
btn_plus=Button(window, text='+', width=3, command=add)
btn_minus=Button(window, text='-', width=3, command=subtract)
btn_mult=Button(window, text='*', width=3, command=multiplication)
btn_div=Button(window, text='/', width=3, command=division)
btn_clear.grid(row=0, column=2, columnspan=4, sticky=W+E+N+S)
btn_plus.grid(row=1, column=2)
btn_minus.grid(row=1, column=3)
btn_mult.grid(row=1, column=4)
btn_div.grid(row=1, column=5)
window.mainloop()
```

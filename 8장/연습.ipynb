```python
infile=open('sales.txt','r')
outfile=open('result.txt','w')

tot=0; count=0

line=infile.readline()
while line != "":
    s=int(line)
    tot+=s; count+=1
    line=infile.readline()

outfile.write('총매출 = '+str(tot)+'\n')
outfile.write('평균 일매출 = '+str(tot/count))

infile.close(); outfile.close()
```

```python
import matplotlib.pyplot as plt
```

```python
X=['Mon','Tue','Wed','Thur','Fri','Sat','Sun']
Y=[15.6,14.2,16.3,18.2,17.1,20.2,22.4]

plt.plot(X,Y)
plt.xlabel('day')
plt.ylabel('temperature')
plt.show()
```

```python
Y1=[15.6,14.2,16.3,18.2,17.1,20.2,22.4]
Y2=[20.1,23.1,23.8,25.9,23.4,25.1,26.3]

plt.plot(X,Y1,X,Y2)
plt.xlabel('day')
plt.ylabel('temperature')
plt.show()
```

```python
plt.plot(X,Y1, label='Seoul')
plt.plot(X,Y2,label='Daegu')
plt.xlabel('day')
plt.ylabel('temperature')
plt.legend(loc='upper left')
plt.title('Temperatures of Cities')
plt.show()
```

```python
years=[1950,1960,1970,1980,1990,2000,2010]
gdp=[67.0,80.0,257.0,1686.0,6505,11865.3,22105.3]

plt.title("GDP per capita")
plt.xlabel('dollars')
plt.ylabel('gdp')

plt.plot(gdp, years, color='red',marker='o', linestyle='dashed')

plt.savefig('gdp_per_capita.png',dpi=600)

plt.show()
```

```python
plt.plot([15.6, 14.2, 16.3, 18.2, 17.1, 20.2, 22.4], 'sb')
plt.show()
```

```python
plt.subplot(1,2,1)
X=['Mon','Tue','Wed','Thur','Fri','Sat','Sun']
Y1=[15.6,14.2,16.3,18.2,17.1,20.2,22.4]
Y2=[20.1,23.1,23.8,25.9,23.4,25.1,26.3]

plt.plot(X,Y1, label='Seoul',color='green',marker='s',linestyle='solid')
plt.xlabel('day')
plt.ylabel('temperature')
plt.legend(loc='upper left')
plt.title('Temperature of Cities')

plt.subplot(1,2,2)
plt.plot(X,Y2, label='Daegu',color='red',marker='o',linestyle='dashdot')
plt.xlabel('day')
plt.ylabel('temperature')
plt.legend(loc='upper left')
plt.title('Temperature of Cities')

plt.show()
```

```python
X=['Mon','Tue','Wed','Thur','Fri','Sat','Sun']
Y=[15.6,14.2,16.3,18.2,17.1,20.2,22.4]
c=['yellow','cyan','pink','purple','green','red','blue']
plt.bar(X,Y,color=c)
plt.title('Weekly weather')
plt.xlabel('day')
plt.ylabel('temperature')
plt.show()
```

```python
plt.barh(X,Y,alpha=0.1)
plt.title('Weekly weather')
plt.xlabel('day')
plt.ylabel('temperature')
plt.show()
```

```python
times=[8,14,2]
timelabels=['Sleep','Study','Play']

plt.pie(times,labels=timelabels,autopct='%.2f')

plt.show()
```
